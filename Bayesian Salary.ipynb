```python
import pandas as pd
import numpy as np
```

```python
salary_train=pd.read_csv('SalaryData_Train (2).csv')
salary_test=pd.read_csv('SalaryData_Test (2).csv')
```

```python
salary_train.head()
```

```python
salary_test.head()
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
##Preprocessing the data. As, there are categorical variables
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
for i in string_columns:
    salary_train[i]= number.fit_transform(salary_train[i])
    salary_test[i]=number.fit_transform(salary_test[i])
```

```python
salary_train
```

```python
salary_test
```

```python
colnames = salary_train.columns
colnames
```

```python
len(colnames)
```

There are a total of 13 columns including 'Salary'

```python
x_train = salary_train[colnames[0:13]]
y_train = salary_train[colnames[13]]
x_test = salary_test[colnames[0:13]]
y_test = salary_test[colnames[13]]
```

```python
## building naive bayes models

from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

```python
## Building Multinomial Naives Bayes model for training data

classifier_mb = MB()
classifier_mb.fit(x_train,y_train)
pred_mb = classifier_mb.predict(x_train)
accuracy_mb_train = np.mean(pred_mb == y_train)
accuracy_mb_train
```

The Multinomial Naives Bayes model running the train data shows accuracy of 77.3%

```python
pd.crosstab(pred_mb, y_train)
```

```python
## Building Multinomial Naives Bayes model for test data

pred_mb_test = classifier_mb.predict(x_test)
accuracy_mb_test = np.mean(pred_mb_test == y_test)
accuracy_mb_test
```

The Multinomial Naives Bayes model running the test data shows accuracy of 77.5%

```python
pd.crosstab(pred_mb_test, y_test)
```

```python
## Building Gaussian Naives Bayes model for training data

classifier_gb = GB()
classifier_gb.fit(x_train, y_train)
pred_gb = classifier_gb.predict(x_train)
accuracy_gb_train = np.mean(pred_gb == y_train)
accuracy_gb_train
```

The Gaussian Naives Bayes model running the train data shows accuracy of 79.5%

```python
pd.crosstab(pred_gb,y_train)
```

```python
## Building Gaussian Naives Bayes model for test data

pred_gb_train = classifier_gb.predict(x_test)
accuracy_gb_test = np.mean(pred_gb_train == y_test)
accuracy_gb_test
```

The Gaussian Naives Bayes model running the test data shows accuracy of 79.46%

```python
pd.crosstab(pred_gb_train,y_test)
```

# We've achieve higher accuracy when using Gaussian Naves Bayes model (79.5% accuracy for training data). Therefore, it is best to use GB instead of MB for this prediction
